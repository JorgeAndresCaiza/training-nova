```python
%sql
SELECT
  job_id,
  COUNT(DISTINCT run_id) AS run_count
FROM
  system.lakeflow.job_run_timeline
WHERE
  result_state IS NOT NULL
GROUP BY
  job_id
HAVING
  COUNT(DISTINCT run_id) > 1
ORDER BY
  run_count DESC
```
